```python
genre_tags = sorted(df_genre.iloc[:, 0].unique())
data = []

for thing in df2.Genres:
    genres = np.zeros(len(genre_tags))

    for i in range(len(genre_tags)):
        col = genre_tags[i]
        if col in thing:
            genres[i] = 1
        
    data.append(genres)
    
df_gf = pd.DataFrame(data, columns=genre_tags)
df_gf.head()
```
